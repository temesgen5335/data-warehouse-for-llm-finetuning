```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

# Set up Selenium webdriver for Firefox
driver = webdriver.Firefox()
#url = "https://www.ethiobookreview.com/book/ye-ethiopia-tarik-ke-lusi-arat-million"
driver.get(url_3)
time.sleep(3)  # Let the page load

# Use BeautifulSoup to parse the page source
soups = BeautifulSoup(driver.page_source, 'html.parser')

# Find the element containing the text data
text_data_element = soups.find('div', class_='et_pb_module et_pb_text et_pb_text_1 et_pb_bg_layout_light')

if text_data_element:
    # Extract text data
    text_data = text_data_element.get_text()
    print(text_data)
else:
    print("Text data element not found")

# Close the Selenium webdriver
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup

#url = "https://am.wikipedia.org/wiki/%E1%8B%A8%E1%8A%A2%E1%89%B5%E1%8B%AE%E1%8C%B5%E1%8B%AB_%E1%8A%A6%E1%88%AD%E1%89%B6%E1%8B%B6%E1%8A%AD%E1%88%B5_%E1%89%B0%E1%8B%8B%E1%88%95%E1%8B%B6_%E1%89%A4%E1%89%B0_%E1%8A%AD%E1%88%AD%E1%88%B5%E1%89%B2%E1%8B%AB%E1%8A%95"

# Fetch the HTML content
response = requests.get(url_5)
content = response.content

# Parse the HTML content
soup5 = BeautifulSoup(content, "html.parser")


# Extract text from paragraphs (adjust selector as needed)
paragraphs = soup5.find_all("p")
all_text = []
for paragraph in paragraphs:
    all_text.append(paragraph.get_text(strip=True))
print(all_text)
```

```python
import re

paragraphs = soup5.find_all("p")
list_paragraphs = []

for paragraph in paragraphs:
    lines = paragraph.get_text(strip=True)
    str_lines = str(lines)
    clean = re.compile('<.*==?>') #compiles a regular expression pattern
    clean2 = (re.sub(clean, '' ,str_lines)) # used the re.sub() method to replace any matches of this pattern in str_cells with an empty string
    list_paragraphs.append(clean2)

print(list_paragraphs)
```

```python
#getting additional data from tables on the website
rows = soup5.find_all('tr')
for row in rows:
    row_td = row.find_all('td')

print(row_td)
```

```python
str_cells = str(row_td)
clean_table = BeautifulSoup(str_cells, 'lxml').get_text()
print(clean_table)
```
