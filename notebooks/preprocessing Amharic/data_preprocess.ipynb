# Character level normalization

```python
import re
#method to normalize character level mismatch such as áŒ¸áˆ€á‹­ and á€áˆá‹­
def normalize_char_level_mismatch(input_token):
    rep1=re.sub('[áˆƒáŠ…áŠƒáˆáˆ“áŠ»]','áˆ€',input_token)
    rep2=re.sub('[áˆ‘áŠá‹…]','áˆ',rep1)
    rep3=re.sub('[áŠ‚áˆ’áŠº]','áˆ‚',rep2)
    rep4=re.sub('[áŠŒáˆ”á‹„]','áˆ„',rep3)
    rep5=re.sub('[áˆ•áŠ…]','áˆ…',rep4)
    rep6=re.sub('[áŠ†áˆ–áŠ¾]','áˆ†',rep5)
    rep7=re.sub('[áˆ ]','áˆ°',rep6)
    rep8=re.sub('[áˆ¡]','áˆ±',rep7)
    rep9=re.sub('[áˆ¢]','áˆ²',rep8)
    rep10=re.sub('[áˆ£]','áˆ³',rep9)
    rep11=re.sub('[áˆ¤]','áˆ´',rep10)
    rep12=re.sub('[áˆ¥]','áˆµ',rep11)
    rep13=re.sub('[áˆ¦]','áˆ¶',rep12)
    rep14=re.sub('[á‹“áŠ£á‹]','áŠ ',rep13)
    rep15=re.sub('[á‹‘]','áŠ¡',rep14)
    rep16=re.sub('[á‹’]','áŠ¢',rep15)
    rep17=re.sub('[á‹”]','áŠ¤',rep16)
    rep18=re.sub('[á‹•]','áŠ¥',rep17)
    rep19=re.sub('[á‹–]','áŠ¦',rep18)
    rep20=re.sub('[áŒ¸]','á€',rep19)
    rep21=re.sub('[áŒ¹]','á',rep20)
    rep22=re.sub('[áŒº]','á‚',rep21)
    rep23=re.sub('[áŒ»]','áƒ',rep22)
    rep24=re.sub('[áŒ¼]','á„',rep23)
    rep25=re.sub('[áŒ½]','á…',rep24)
    rep26=re.sub('[áŒ¾]','á†',rep25)
    #Normalizing words with Labialized Amharic characters such as á‰ áˆá‰±á‹‹áˆ or  á‰ áˆá‰±áŠ áˆ to  á‰ áˆá‰·áˆ  
    rep27=re.sub('(áˆ‰[á‹‹áŠ ])','áˆ',rep26)
    rep28=re.sub('(áˆ™[á‹‹áŠ ])','áˆŸ',rep27)
    rep29=re.sub('(á‰±[á‹‹áŠ ])','á‰·',rep28)
    rep30=re.sub('(áˆ©[á‹‹áŠ ])','áˆ¯',rep29)
    rep31=re.sub('(áˆ±[á‹‹áŠ ])','áˆ·',rep30)
    rep32=re.sub('(áˆ¹[á‹‹áŠ ])','áˆ¿',rep31)
    rep33=re.sub('(á‰[á‹‹áŠ ])','á‰‹',rep32)
    rep34=re.sub('(á‰¡[á‹‹áŠ ])','á‰§',rep33)
    rep35=re.sub('(á‰¹[á‹‹áŠ ])','á‰¿',rep34)
    rep36=re.sub('(áˆ[á‹‹áŠ ])','áŠ‹',rep35)
    rep37=re.sub('(áŠ‘[á‹‹áŠ ])','áŠ—',rep36)
    rep38=re.sub('(áŠ™[á‹‹áŠ ])','áŠŸ',rep37)
    rep39=re.sub('(áŠ©[á‹‹áŠ ])','áŠ³',rep38)
    rep40=re.sub('(á‹™[á‹‹áŠ ])','á‹Ÿ',rep39)
    rep41=re.sub('(áŒ‰[á‹‹áŠ ])','áŒ“',rep40)
    rep42=re.sub('(á‹°[á‹‹áŠ ])','á‹·',rep41)
    rep43=re.sub('(áŒ¡[á‹‹áŠ ])','áŒ§',rep42)
    rep44=re.sub('(áŒ©[á‹‹áŠ ])','áŒ¯',rep43)
    rep45=re.sub('(áŒ¹[á‹‹áŠ ])','áŒ¿',rep44)
    rep46=re.sub('(á‰[á‹‹áŠ ])','á',rep45)
    rep47=re.sub('[á‰Š]','á‰',rep46) #á‰ can be written as á‰Š
    rep48=re.sub('[áŠµ]','áŠ©',rep47) #áŠ© can be also written as áŠµ  
    return rep48
```

```python
df_am['Unnamed: 0'] = df_am['Unnamed: 0'].str.replace('[^\w\s]','')

df_am['Unnamed: 0'] = df_am['Unnamed: 0'].apply(lambda x: normalize_char_level_mismatch(x))
```

```python
def count_words(text):
    if pd.isna(text):
        return 0
    return len(text.split())

df_am['word_count'] = df_am['Unnamed: 0'].apply(count_words)
df_am.head()
```

```python
import re
def extract_tags(text):
    tags = re.findall(r'#\w+', text)
    return ','.join(tags)

df_am['tags'] = df_am['Unnamed: 0'].apply(extract_tags)
```
